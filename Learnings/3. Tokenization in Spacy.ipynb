<h3>Customizing tokenizer</h3>

```python
#when having mixed words

doc = nlp_spacy("gimme double cheese extra large healthy pizza")

tokens = [token.text for token in doc]
tokens
```

here gimme is two word

```python
from spacy.symbols import ORTH

nlp_spacy.tokenizer.add_special_case("gimme", [
    {ORTH: "gim"}, #we can modify it into give
    {ORTH: "me"} ])


doc = nlp_spacy("gimme double cheese extra large healthy pizza")

tokens = [token.text for token in doc]
tokens
```
